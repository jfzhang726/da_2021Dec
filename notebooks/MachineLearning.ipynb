```python
DICT_SUBURB_COUNCIL = {}
for idx, row in df[~df['CouncilArea'].isna()].iterrows():
    values = DICT_SUBURB_COUNCIL.get(row['Suburb'], set())
    values.add(row['CouncilArea'])
    DICT_SUBURB_COUNCIL[row['Suburb']] = values

DICT_SUBURB_COUNCIL
```

```python
count_unmatched, count_matched = 0, 0
for idx, row in df[df['CouncilArea'].isna()].iterrows():
    if row['Suburb'] not in DICT_SUBURB_COUNCIL:
        print('unmatched ', row['Suburb'])
        count_unmatched += 1
    else:
        #print('matched ', row['Suburb'], DICT_SUBURB_COUNCIL[row['Suburb']])
        count_matched += 1
print('unmatched {} matched {}'.format(count_unmatched, count_matched))
```

```python
def fill_missing_councilarea(suburb, council_area):
    if council_area is not np.nan:
        return council_area
    council_area = DICT_SUBURB_COUNCIL.get(suburb, None)
    if council_area is None:
        return None
    return list(council_area)[0]
df['CouncilArea'] = df.apply(lambda row: fill_missing_councilarea(row['Suburb'], row['CouncilArea']), axis=1)
```

Prepare training and testing data.

```python
from sklearn.model_selection import train_test_split


X_df = df[["Suburb", "Address", "Rooms", "Type", "Method", "SellerG", "Date", "Distance", "Postcode", "Bedroom2", "Bathroom", "Car", "Landsize", "CouncilArea", "Regionname", "Propertycount"]]
y_df = df[["Price"]]

X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, test_size=0.3, random_state=0)
```

```python
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from sklearn.compose import ColumnTransformer


# encoders fit on entire dataset
suburb_encoder = OneHotEncoder(sparse=False)
suburb_encoder.fit(X_df['Suburb'].to_numpy().reshape((-1,1)))

type_encoder = OneHotEncoder(sparse=False)
type_encoder.fit(df['Type'].to_numpy().reshape(-1,1))

method_encoder = OneHotEncoder(sparse=False)
method_encoder.fit(df['Method'].to_numpy().reshape(-1,1))

sellerg_encoder = OneHotEncoder(sparse=False)
sellerg_encoder.fit(df['SellerG'].to_numpy().reshape(-1,1))

postcode_encoder = OneHotEncoder(sparse=False)
postcode_encoder.fit(df['Postcode'].to_numpy().reshape(-1,1))

councilarea_encoder = OneHotEncoder(sparse=False)
councilarea_encoder.fit(df['CouncilArea'].to_numpy().reshape(-1,1))

regionname_encoder = OneHotEncoder(sparse=False)
regionname_encoder.fit(df['Regionname'].to_numpy().reshape(-1,1))


# standardization on training set
MINMAX_COLUMNS = ["Rooms", 
                  "Distance", 
                  "Bedroom2", 
                  "Bathroom", 
                  "Car", 
                  "Landsize", 
                  #"Propertycount"
                 ]
minmax_scaler = MinMaxScaler()
minmax_scaler.fit(X_train[MINMAX_COLUMNS].to_numpy())




def get_vectors(data_df, target_df):
    suburb_vectors = suburb_encoder.transform(data_df['Suburb'].to_numpy().reshape(-1,1))
    type_vectors = type_encoder.transform(data_df['Type'].to_numpy().reshape(-1,1))
    method_vectors = method_encoder.transform(data_df['Method'].to_numpy().reshape(-1,1))
    sellerg_vectors = sellerg_encoder.transform(data_df['SellerG'].to_numpy().reshape(-1,1))
    postcode_vectors = postcode_encoder.transform(data_df['Postcode'].to_numpy().reshape(-1,1))
    councilarea_vectors = councilarea_encoder.transform(data_df['CouncilArea'].to_numpy().reshape(-1,1))
    regionname_vectors = regionname_encoder.transform(data_df['Regionname'].to_numpy().reshape(-1,1))
    
    numerical_vectors = minmax_scaler.transform(data_df[MINMAX_COLUMNS].to_numpy())
    
    
    
    
    
    X_vectors = np.hstack((suburb_vectors, 
                           type_vectors,
                           method_vectors,
                           sellerg_vectors,
                           postcode_vectors,
                           councilarea_vectors,
                           regionname_vectors,
                           numerical_vectors
                          ))
    y = target_df.to_numpy().reshape(-1)
    return X_vectors, y

X_train_vec, y_train_vec = get_vectors(X_train, y_train)
X_test_vec, y_test_vec = get_vectors(X_test, y_test)
```

```python
from sklearn import linear_model
from sklearn.ensemble import BaggingRegressor
from sklearn.model_selection import cross_validate


elastic_net_model = linear_model.ElasticNet(random_state=0)
result = cross_validate(elastic_net_model, X_train_vec, y_train_vec, cv=5, scoring=['r2', 'neg_mean_absolute_error', 'neg_mean_absolute_percentage_error'])
print(result)
#elastic_net.fit(X_train_vec, y_train_vec)

sgd_model = linear_model.SGDRegressor(max_iter=1000, tol=1e-3)
result = cross_validate(sgd_model, X_train_vec, y_train_vec, cv=5, scoring=['r2', 'neg_mean_absolute_error', 'neg_mean_absolute_percentage_error'])
print(result)

```

```python
from sklearn import linear_model
from sklearn.ensemble import BaggingRegressor, AdaBoostRegressor
from sklearn.model_selection import cross_validate

bagging_elastic_net = BaggingRegressor(base_estimator=linear_model.ElasticNet(),
                                       n_estimators=10,
                                       random_state=0)

result = cross_validate(bagging_elastic_net, X_train_vec, y_train_vec, cv=5, scoring=['r2', 'neg_mean_absolute_error', 'neg_mean_absolute_percentage_error'])
print(result)
```

```python
adaboost_decision_tree = AdaBoostRegressor(random_state=0, n_estimators=100)
result = cross_validate(adaboost_decision_tree, X_train_vec, y_train_vec, cv=5, scoring=['r2', 'neg_mean_absolute_error', 'neg_mean_absolute_percentage_error'])
print(result)

adaboost_elasticnet = AdaBoostRegressor(base_estimator=linear_model.ElasticNet(random_state=0), random_state=0, n_estimators=100)
result = cross_validate(adaboost_elasticnet, X_train_vec, y_train_vec, cv=5, scoring=['r2', 'neg_mean_absolute_error', 'neg_mean_absolute_percentage_error'])
print(result)
```

```python
from sklearn import metrics
def predict_evaluate(model, X_test, y_test):
    predicts = model.predict(X_test)
    mae = metrics.mean_absolute_error(y_true=y_test, y_pred=predicts)
    mse = metrics.mean_squared_error(y_true=y_test, y_pred=predicts)
    rmse = metrics.mean_squared_error(y_true=y_test, y_pred=predicts, squared=False)
    r2 = metrics.r2_score(y_true=y_test, y_pred=predicts)
    mape = metrics.mean_absolute_percentage_error(y_true=y_test, y_pred=predicts)
    return mae, mse, rmse, r2, mape

elastic_net_model.fit(X_train_vec, y_train_vec)
mae, mse, rmse, r2, mape = predict_evaluate(elastic_net_model, X_test_vec, y_test_vec)
print('MAE:', mae)
print('MSE:', mse)
print('RMSE:', rmse)
print('R2:', r2)
print('Mean Absolute Precentage Error:', mape)
```
